```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
from io import StringIO

%matplotlib inline
```

```python
sheets = pd.read_excel('/content/Vivino-export.xlsx', sheet_name=None, header=None, engine="openpyxl")

dfs = []
for sheet, raw in sheets.items():
    if raw.empty:
        continue

    text = "\n".join(raw.iloc[:, 0].dropna().astype(str))

    part = pd.read_csv(StringIO(text), sep=",", engine="python", on_bad_lines="skip")

    part.columns = ["name", "country", "region", "rating", "rating_count", "price"]
    part["sheet"] = sheet
    dfs.append(part)

df = pd.concat(dfs, ignore_index=True)
df.head()
```

```python
df["year"] = df.name.str.extract(r"(\d{4})")
df
```

```python
df.name = df.name.replace(r"\s*\d{4}$", "", regex=True)
df
```

```python
BAD_CHARS = r"[()'`,√´©¢®¬∞≠∫Æ¨±¥Äô]"
```

```python
names_to_clean = ["country", "region", "name"]

for naam in names_to_clean:
    df[naam] = (df[naam].astype(str).str.replace(BAD_CHARS, "", regex=True).str.strip())
```

```python
df["price"] = pd.to_numeric(df["price"], errors="coerce").round(2)
df
```
