```python
sheets = pd.read_excel("Vivino-export.xlsx", engine="openpyxl", sheet_name=None)

dfs = []
bad_lines_log = []

for sheet_name, df_raw in sheets.items():
    if df_raw.empty:
        continue

    text = "\n".join(df_raw.iloc[:, 0].astype(str).tolist())

    try:
        df_part = pd.read_csv(
            StringIO(text),
            sep=",",
            engine="python",
            quoting=csv.QUOTE_MINIMAL,
            on_bad_lines="skip"
        )
    except Exception as e:
        bad_lines_log.append((sheet_name, f"FAILED ENTIRE SHEET: {e}"))
        continue

    df_part.columns = ["name", "country", "region", "rating", "rating_count", "price"]
    df_part["sheet"] = sheet_name

    dfs.append(df_part)

df = pd.concat(dfs, ignore_index=True)
```
